```python
import math
import networkx as nx
import matplotlib.pyplot as plt
```

```python
class Graph:
    def __init__(self, nodes, edges, nodes_infos):
        self.nodes = nodes
        self.edges = edges
        self.nodes_infos = nodes_infos
        self.adjacency_list = self.build_adjacency_list()

    def build_adjacency_list(self):
        adjacency_list = {node: [] for node in range(len(self.nodes))}
        for edge in self.edges:
            node1, node2, distance = edge
            adjacency_list[node1].append((node2, distance))
            adjacency_list[node2].append((node1, distance))
        return adjacency_list
    
    def viz(self):
        # Create a NetworkX graph object from the Graph class object
        G = nx.Graph()
        for node in self.nodes:
            G.add_node(node)
        for edge in self.edges:
            G.add_edge(edge[0], edge[1], weight=edge[2])
        # Create a dictionary of node colors based on whether the node is in l_init or not
        node_colors = {node: 'b' for node in self.nodes}
        # Create a list of edge colors based on whether the edge is in l_gw or not
        edge_colors = ['g' for edge in self.edges]
        # Calculate the spring layout of the graph to minimize edge crossings
        pos = nx.spring_layout(G, k=0.5, iterations=100000)
        # Draw the graph
        nx.draw_networkx_nodes(G, pos, node_size=300, node_color=list(node_colors.values()))
        nx.draw_networkx_edges(G, pos, edgelist=self.edges, edge_color=edge_colors)
        nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif')
        plt.axis('off')
        plt.show()
```

```python
l_nodes = [0, 1, 2, 3, 4, 5, 6, 7, 8]
l_edges = [[0, 1, 3], [0, 2, 3], [0, 3, 2], [0, 4, 2], [0, 5, 2], [0, 6, 2], [0, 7, 6], [0, 8, 6], [1, 2, 7], [1, 3, 1], [1, 4, 7], [1, 5, 4], [1, 6, 4], [1, 7, 4], [1, 8, 10], [2, 3, 7], [2, 4, 1], [2, 5, 4], [2, 6, 4], [2, 7, 10], [2, 8, 4], [3, 4, 6], [3, 5, 2], [3, 6, 5], [3, 7, 2], [3, 8, 10], [4, 5, 5], [4, 6, 2], [4, 7, 10], [4, 8, 2], [5, 6, 5], [5, 7, 4], [5, 8, 9], [6, 7, 9], [6, 8, 4], [7, 8, 14]]
```
